예제 4.6
========

-이진탐색(반복구조)-

### 1.1 문제 정의

* 이스트에 n개의 항목이 들어있다. 이 리스트에서 '탐색키'를 가진 항목을 찾아라. 단, 리스트의 항목들은 정렬되어 있다.

### 1.2 알고리즘 설명

* 실행 효율적인 측면에서는 순환구조 보다는 반복구조가 유리하다.
1. 리스트의 중앙에 있는 항목을 먼저 조사한다.
2. 만약 항목이 탐색키보다 크다면, 찾는 항목은 왼쪽에 있고, 오른 쪽은 더이상 조사할 필요가 없음.
3. 반대면 답은 오른 쪽에 있고 왼쪽은 검사할 필요가 없다.
4. 따라서, 각 단계마다 검사할 범위가 즉, 문제의 크기가 반으로 줄어든다.

### 1.3 손으로 푼 예제

### 1.4 코드 설명

* 입력변수: 정렬된 리스트
* 출력변수: 찾는 값의 리스트에서의 위치
* 함수: binary_search_iter()

1. 가장 낮은 값과 가장 높은 값의 평균을 중앙값으로 한다.
2. 중앙값과 key값이 같다면 탐색에 성공
3. 중앙값보다 key값이 크다면 가장 작은 값 대신 중앙값에 +1한 값을 대입한다.
4. 중앙값이 key값보다 크다면 가장 큰 값 대신 중앙값에 -1한 값을 대입한다.
5. 2~4번을 반복한다.

### 1.5 알고리즘 코드

```python
def binary_search_iter(A,key,low,high):
    while(low<=high):
        mid = (low+high)//2
        if key==A[mid]:
            return mid
        elif key>A[mid]:
            low = mid+1
        else:
            high = mid - 1
    return -1
```

### 1.6 테스트 코드

```python
listA=[1,3,8,13,13,16,21,26,28,34,38,39,40,44,49]
print("입력 리스트=", listA)
print("33탐색 --> ", binary_search_iter(listA, 33, 0, len(listA)-1))
print("16탐색 --> ", binary_search_iter(listA, 16, 0, len(listA)-1))
```
