예제 4.4
============

-위상정렬(축소 정복 기법)-

### 1.1 문제 정의

- 방향 그래프 G = (V,E)가 주어졌다. G에 존재하는 각 정점들의 선행 순서를 위배하지 않으면서 모든 정점들을 순서대로 나열하라. 단, 숙소 정복 기법을 사용하여 알고리즘을 구성하여라.

### 1.2 알고리즘 설명

1. 모든 정점의 진입차수를 구한 후, 진입차수가 0인 정점들을 리스트에 저장하고 리스트가 공백이 아닐 때까지 진입차수가 0인 정점을 하나 꺼낸 후, 연결된 정점들의 진입차수를 감소시킨다.

* 축소 정복 기법의 알고리즘
1. 방향 그래프에서 진입차수가 0인 정점(선수과목이 없어 바로 수강할 수 있는 과목) 중에서 하나를 임의로 선택한다.
2. 선택된 정점을 삭제한다. 정점이 삭제되므로 이 정점에서 진출하는 모든 간선들도 삭제한다. 이떄, 간선이 삭제되므로 삭제되는 간선과 연결된 이웃 정점들의 진입 차수는 갱신되어야 한다.
3. 이 과정을 모든 정점이 삭제될 때까지 반복한다. 정점이 삭제되는 순서가 위상순서이다.

### 1.3 손으로 푼 예제

### 1.4 코드 설명

*입력변수: 인접리스트
*출력변수: 위상순서
*함수설명: topological_sort()

1. 방향그래프를 인접리스트로 입력받는다.
2. 집입차수가 0인 정점 리스트를 만든다.
3. 리스트가 공백이 아닐 때까지 진입차수가 0인 정점을 하나씩 꺼낸다.
4. 진입차수가 0이면 visit에 저장하고 이후, 위상 순서를 출력한다.

### 1.5 알고리즘 코드

```python
def topological_sort(graph):
    inDeg = {}
    for v in graph:
        inDeg[v]=0
    for v in graph:
        for u in graph[v]:
            inDeg[u]+=1

    vlsit = []
    for v in graph:
        if inDeg[v]==0:
            vlsit.append(v)

    while vlsit:
        v=vlsit.pop()
        print(v, end=' ')

        for u in graph[v]:
            inDeg[u] -= 1
            if inDeg[u] == 0:
                vlsit.append(u)
```

### 1.6 테스트 코드

```python
graph={"A":{"B","C","D"}, "B":{"E"}, "C":{"E"},"D":{"F"},"E":{"F"},"F":{ }}
print('topological_sort: ')
topological_sort(graph)
print()
```

### 1.8 복잡도 분석

*O(n+e)
1. 2~7행: 3~4번 행은 O(n)이나, 7행이 간선의 개수 만큼 처리해야 하므로, 전체 복잡도는 O(n+e)이다.
2. 9~12행: 10행의 반복문은 정점의 수만큼 반복되므로 O(n)이다.
3. 14~21행: 최악의 경우에, 전체 간선의 수 만큼 수행해야 하므로 O(e)이다.
4. 전체 복잡도는 O(n+e)+O(n)+O(e)=O(n+e)이다.
