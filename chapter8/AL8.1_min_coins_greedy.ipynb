# 알고리즘 8.1  
-거스름돈 최소화 알고리즘-

### 1.1 문제 정의

액면가가 서로 다른 m가지의 동전이 있다. 거스름돈으로 V원을 동전으로만 돌려주어야 한다면 최소 몇 개의 동전이 필요한지를 구하여라.
* 단, 모든 동전은 무한히 사용할 수 있고, 동전들은 액수가 큰 것부터 내림차순으로 순서대로 정렬되어있다.

### 1.2 알고리즘 설명

* 탐욕적 전략: 가장 액수가 큰 동전부터 최대한 사용하면서 거스름돈을 맞춤
1. 어떤 결정을 해야 할 때 마다 "그 순간에 최적"이라고 생각되는 것을 선택히거 그 답이 최적의 해답을 주는지 검증해야 한다.
2. 탐욕적 기법이 사용될 수 있는 경우  
 (1) 탐욕적 기법을 사용해도 항상 최적해를 구할 수 있는 문제  
 (2) 최적해를 구하는 것이 현실적으로 불가능할 경우  
3. 알고리즘 설계 과정  
 (1) 현재 주어진 문제에 대한 가장 탐욕적인 해를 선택한다.
 (2) 이 해가 가능한 해인지를 검사한다. 만약, 가능하지 않다면 (1)로 돌아가 다음으로 가장 탐욕적인 해를 선택한다.  
 (3) 만약 가능한 해라면, 문제가 해결되었는지를 검사한다. 만약 아직 문제가 해결되지 않았으면 이 해를 적용해 크기가 줄어든 부분 문제에 대해 다시 (1)로 돌아가 해를 구한다.

### 1.4 코드 설명

* 입력변수: 정렬된 리스트, 잔돈 저장하는 변수
* 출력 변수: 거스름돈으로 낼 수 있는 동전의 개수
* 함수: min_coins_greedy()

1. 함수는 각 동전의 금액을 포함하는 리스트인 coins와 잔돈을 저장하는 변수 V로 입력받는다.
2. 함수는 각 동전의 개수를 저장할 빈 리스트인 count를 만든다.
3. for 루프는 동전 리스트를 순회합니다. 잔돈에서 V를 나눈 몫을 구하여 해당 동전의 개수를 계산한다.
4. 이 개수를 count 리스트에 추가한 후 V에서 해당 동전을 만들기 위해 사용한 금액을 빼준다.
5. 마지막으로 count 리스트를 반환한다.

### 1.5 알고리즘 코드

```python
def min_coins_greedy(coins, V):
    count = []
    for i in range(len(coins)):
        cnt = V // coins[i]
        count.append(cnt)
        V-=cnt*coins[i]
    return count
```

### 1.6 테스트 코드

```python
coins = [500, 100, 50, 10]
V=450
print("잔돈 액수 = ", V)
print("동전 종류 = ", coins)
print("동전 개수 = ", min_coins_greedy(coins, V))
```
