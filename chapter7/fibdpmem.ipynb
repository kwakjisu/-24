# 1.5 알고리즘 코드

```python
def fib_dp_mem(n):
    if(mem[n] == None):
        if n<2:
            mem[n] = n
        else:
            mem[n] = fib_dp_mem(n-1)+fib_dp_mem(n-2)
    return mem[n]
```

# 1.6 실행 코드

```python
n=8
mem=[None]*(n+1)
print('동적계획(메모이제이션): Fibonacci(%d) = '%n, fib_dp_mem(n))
```
